Написать класс нейрон, который принимает вектор X, предсказывает y. Нейрон имеет пороговую функцию активации (слайд 38). Используйте алгоритм со слайда 46. Не забывайте про применение смещения (порога, bias) слайд 39.

Обучите и проверьте работоспособность нейрона на логических функциях И, ИЛИ, исключающее ИЛИ Векторами X будут все возможные сочетания 0 и 1

0 1

1 0

1 1

0 0

Ответы для И будут (0, 0, 1, 0), для ИЛИ (1, 1, 1, 0), для исключающее ИЛИ (1, 1, 0, 0)

```python
import random
import math

class Neuron:
    def __init__(self, input_size):
        self.weights = [random.uniform(-1, 1) for _ in range(input_size)]
        self.bias = random.uniform(-1, 1)

    def predict(self, X):
        logits = sum([X[i]*self.weights[i] for i in range(len(X))]) + self.bias
        y_predict = self.activation(logits)
        return y_predict

    def activation(self, logits):
        if logits > 0:
            return 1
        else:
            return 0

    def train(self, X, y, max_iterations=1000):
        perfect = False
        iteration = 0
        while not perfect and iteration < max_iterations:
            perfect = True
            for i in range(len(X)):
                X_i = X[i]
                y_i = y[i]
                y_predict = self.predict(X_i)
                if y_predict != y_i:
                    perfect = False
                    if y_predict == 0:
                        self.weights = [self.weights[j] + X_i[j] for j in range(len(X_i))]
                        self.bias += 1
                    else:
                        self.weights = [self.weights[j] - X_i[j] for j in range(len(X_i))]
                        self.bias -= 1
            iteration += 1

    def get_weights(self):
        return self.weights, self.bias
```

#Проверим, как это дело работает на логических функциях.

Логическое И. 

Тут всё норм. Тут всё нравится.

```python
X_and = [[0, 1], [1, 0], [1, 1], [0, 0]]
y_and = [0, 0, 1, 0]
neuron_and = Neuron(input_size=2)
neuron_and.train(X_and, y_and)
print([int(neuron_and.predict(X_and[i])) for i in range(len(X_and))])
```

Логическое ИЛИ. 

Тут тоже всё норм. Тут тоже всё нравится.

```python
X_or = [[0, 1], [1, 0], [1, 1], [0, 0]]
y_or = [1, 1, 1, 0]
neuron_or = Neuron(input_size=2)
neuron_or.train(X_or, y_or)
print([int(neuron_or.predict(X_or[i])) for i in range(len(X_or))])
```

Логическое исключающее ИЛИ. 

Тут вообще не норм. Тут вообще не нравится.

```python
X_xor = [[0, 1], [1, 0], [1, 1], [0, 0]]
y_xor = [1, 1, 0, 0]
neuron_xor = Neuron(input_size=2)
neuron_xor.train(X_xor, y_xor)
print([int(neuron_xor.predict(X_xor[i])) for i in range(len(X_xor))])
```

Почему так? Потому что, скорее всего, это связано с тем, что для решения задачи XOR недостаточно одного нейрона. Тут либо использовать несколько нейронов. Либо переходить от бинарной функции активации к какой-то другой.
